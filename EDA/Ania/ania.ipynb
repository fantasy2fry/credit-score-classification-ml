```python
df = pd.read_csv('../../data/credit_score.csv')
df
```

```python
df.describe()
```

```python
df_short = df[["INCOME","DEBT", "SAVINGS", "R_EDUCATION", "CAT_GAMBLING", "DEFAULT"]]
```

wydawały się ważne te kolumny to chciałam je zbadać

```python
fig, axs = plt.subplots(2, 3, figsize=(16, 8))
axs = axs.flatten()
for i, col in enumerate(df_short.columns[:-1]):
    sns.scatterplot(data=df_short, x=col, y=df_short.groupby(col)['DEFAULT'].mean(), ax=axs[i])
plt.suptitle("Target variable vs. features")
plt.tight_layout()
```

```python

sns.boxplot(data=df_short, y='INCOME', x='DEFAULT')
```

```python
average_default = df_short.groupby('CAT_GAMBLING')['DEFAULT'].mean()
print(average_default)
```

im wyższe cat_gambling, tym większa szansa na default=1

```python
bins = [0, 10000, 30000, 60000, 100000, 150000, 662094]

df_short['income_cat'] = pd.cut(df_short['INCOME'], bins=bins)

average_default = df_short.groupby('income_cat')['DEFAULT'].mean()
print(average_default)
```

wydaje się,że zarobki niebardzo wplywają na spłacanie kredytu, ale nie rozważyłam możliwości, że im więcej ktoś zarabia, tym większy kredyt bierze. Może być tak, że małe kredyty są częściej spłacane przez osoby, które więcej zarabiają, ale ci mniej zarabiający nie biorą wysokich kredytów, które możliwe, że są spłacane ,,gorzej"

```python
average_default = df_short.groupby('DEFAULT')['SAVINGS'].mean()
print(average_default)
```

```python
average_default = df_short.groupby('DEFAULT')['INCOME'].mean()
print(average_default)
```

no właśnie:)

```python
low_debt = df_short[df_short['DEBT'] <= 5.396675e+04]


average_default = low_debt.groupby('income_cat')['DEFAULT'].mean()
print(average_default)

average_debt = df_short[(df_short['DEBT'] > 5.396675e+04) & (df_short['DEBT'] <= 3.950955e+05)]

average_default2 = average_debt.groupby('income_cat')['DEFAULT'].mean()
print(average_default2)

high_debt = df_short[(df_short['DEBT'] > 3.950955e+05) & (df_short['DEBT'] <= 1.193230e+06)]

average_default2 = high_debt.groupby('income_cat')['DEFAULT'].mean()
print(average_default2)
highest_debt = df_short[df_short['DEBT'] > 1.193230e+06]

average_default2 = highest_debt.groupby('income_cat')['DEFAULT'].mean()
print(average_default2)
```

tak jak przypuszczałam, osoby z małymi zarobkami nie biorą bardzo dużych kredytów. przy podziale na przedziały wartości kredytów już widać, że zarobki mają duży wpływ na spłacanie kredytów: im niższe zarobki, tym więcej niespłaconych kredytów

```python
df = df.iloc[:, 1:]
mapping = {'No': 0, 'Low': 1, 'High': 2}
df['CAT_GAMBLING'] = df['CAT_GAMBLING'].map(lambda x: mapping[x])

corr = df.corr()
plt.figure(figsize=(20, 20))
sns.heatmap(corr, annot=False, cmap='coolwarm', fmt=".2f")
plt.show()
```
